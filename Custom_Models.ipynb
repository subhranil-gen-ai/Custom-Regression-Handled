# Custom Linear Regression

```python
from sklearn.base import BaseEstimator,TransformerMixin
```

```python
class MyLinearRegression(BaseEstimator,TransformerMixin):
  def __init__(self,**kwargs):
    pass
  def fit(self,X,y):
    X_b=np.c_[np.ones((len(X),1)),X]
    self.theta_best=np.linalg.inv(X_b.T @ X_b)@X_b.T@y
    self.intercept=self.theta_best[0]
    self.coefficient=self.theta_best[1:]
    return self
  def predict(self,X):
     X_b=np.c_[np.ones((len(X),1)),X]
     y_pred=X_b@self.theta_best
     return y_pred
```

```python
custom_lin_reg=MyLinearRegression()
```

```python
custom_lin_reg.fit(X,y)
```

# Custom BGD

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
class CustomBGDRegressor(BaseEstimator, TransformerMixin):
    def __init__(self, eta=0.1, n_iterations=1000, random_state=42):
        self.eta = eta
        self.n_iterations = n_iterations
        self.random_state = random_state
        np.random.seed(self.random_state)
    def fit(self, X_scaled, y):
        m, n = X_scaled.shape
        X_b = np.c_[np.ones((len(X_scaled), 1)), X_scaled]
        self.theta = np.random.randn(n + 1, 1)
        for iteration in range(self.n_iterations):
            gradients = 2 / m * X_b.T @ (X_b @ self.theta - y)
            self.theta = self.theta - self.eta * gradients
        return self
    def predict(self, X_scaled):
        X_b = np.c_[np.ones((len(X_scaled), 1)), X_scaled]
        y_pred = X_b @ self.theta
        return y_pred
```

# Custom SGD

```python
class CustomSGDRegressor(BaseEstimator, TransformerMixin):
    def __init__(self, eta=0.1, n_iterations=1000, random_state=42):
        self.eta = eta
        self.n_iterations = n_iterations
        self.random_state = random_state
        np.random.seed(self.random_state)
    def fit(self, X_scaled, y):
        m, n = X_scaled.shape
        X_b = np.c_[np.ones((len(X_scaled), 1)), X_scaled]
        self.theta = np.random.randn(n + 1, 1)
        for iteration in range(self.n_iterations):
           for i in range(m):
              X_i=X_b[i:i+1]
              y_i=y[i:i+1]
              gradients = 2 * X_i.T @ (X_i @ self.theta - y_i)
              self.theta = self.theta - self.eta * gradients
        return self
    def predict(self, X_scaled):
        X_b = np.c_[np.ones((len(X_scaled), 1)), X_scaled]
        y_pred = X_b @ self.theta
        return y_pred
```

```python
custom_SGD=CustomSGDRegressor()
```
